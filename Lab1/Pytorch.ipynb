# BTVN 02

Tạo một tensor ban đầu có giá trị là 2

Định nghĩa hàm số và tính gradient 

y = x^3 + 2x^2 + 5x + 1

Hãy tính dy/dx tại giá trị của x

Dùng phương pháp gradient descent với learning rate alpha = 0.1 để cập nhập giá trị x trong 10 vòng lặp


```python
import torch

# Khởi tạo x = 2
x = torch.tensor(2.0, requires_grad=True)

# Định nghĩa hàm y
y = x**3 + 2*x**2 + 5*x + 1

# Tính đạo hàm dy/dx
y.backward()

print("Giá trị y:", y.item())
print("dy/dx tại x = 2:", x.grad.item())
```

```python
# Reset x
x = torch.tensor(2.0, requires_grad=True)

alpha = 0.1

for i in range(10):
    y = x**3 + 2*x**2 + 5*x + 1
    y.backward()
    
    with torch.no_grad():
        x -= alpha * x.grad   # cập nhật x
    
    x.grad.zero_()  # reset gradient
    
    print(f"Vòng {i+1}: x = {x.item():.6f}")
```

# BTVN 03

Tạo 1 tập dữ liệu giả lập với x là số giờ học ngẫu nhiên từ 1 - 10 và y là số điểm được tính theo công thức y = 3x + 5 + noise
Với noise là một giá trị ngẫu nhiên nhỏ

1. khởi tạo tham số w và b ngẫu nhiên với requieres_grad = True
1. tính MSE
1. Tính gradient
1. Cập nhật tham số w và b bằng gradient descent với learning rate alpha = 0.01
1. Lặp lại quá trình trên trong 100 vòng lặp để xem sự hội tụ của mô hình

```python
import torch

# Fix seed để dễ quan sát hội tụ
torch.manual_seed(0)

# Số mẫu
N = 50

# Dữ liệu
x = torch.rand(N) * 9 + 1          # [1, 10]
noise = torch.randn(N) * 0.5       # nhiễu nhỏ
y = 3 * x + 5 + noise

w = torch.randn(1, requires_grad=True)
b = torch.randn(1, requires_grad=True)

def predict(x):
    return w * x + b

def mse_loss(y_pred, y_true):
    return ((y_pred - y_true) ** 2).mean()

alpha = 0.01
epochs = 100

for epoch in range(epochs):
    # Forward
    y_pred = predict(x)
    loss = mse_loss(y_pred, y)
    
    # Backward: tính gradient
    loss.backward()
    
    # Cập nhật tham số
    with torch.no_grad():
        w -= alpha * w.grad
        b -= alpha * b.grad
    
    # Reset gradient
    w.grad.zero_()
    b.grad.zero_()
    
    # In tiến trình
    if (epoch + 1) % 10 == 0:
        print(f"Epoch {epoch+1:3d} | Loss = {loss.item():.4f} | w = {w.item():.4f} | b = {b.item():.4f}")
```
