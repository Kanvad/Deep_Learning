```python
# BTVN 01
# cho ma trận = [[99,99,99],
#                 [99,99,99],
#                 [99,99,99]]

# hãy tưởng tượng ô 99 trống
# giả sử 1 là x
# giả sử 0 là o

# lấy đầu vào từ phía x và phía o luân phiên
# nếu đầu vào từ phía x là (0,0)
# ma trận sẽ trở thành
# [[x,99,99],
# [99,99,99],
# [99,99,99]]

# nếu đầu vào từ phía o là (0,0)
# thông báo họ nhập lại 
# nếu không thì điền vào ma trận

# thử thách: nếu người đầu tiên có 3 phần tử liên tiếp thì dừng trò chơi
```

```python
import numpy as np

board = np.full((3,3), 99, dtype=int)

def print_board(b):
    sym = {99:'.', 1:'X', 0:'O'}
    for r in range(3):
        print(' '.join(sym[int(v)] for v in b[r]))
    print()

def is_winner(b, p):
    # hàng
    if np.any(np.all(b == p, axis=1)): return True
    # cột
    if np.any(np.all(b == p, axis=0)): return True
    # chéo chính
    if np.all(np.diag(b) == p): return True
    # chéo phụ
    if np.all(np.diag(np.fliplr(b)) == p): return True
    return False

def valid_move(b, r, c):
    return 0 <= r < 3 and 0 <= c < 3 and b[r, c] == 99

def play():
    current = 1  # X bắt đầu
    moves = 0
    while True:
        print_board(board)
        player = 'X' if current == 1 else 'O'
        try:
            s = input(f"Người {player} nhập hàng,cột (ví dụ 0 0): ").strip()
            r, c = map(int, s.split())
        except:
            print("Nhập không hợp lệ. Thử lại.")
            continue
        if not (0 <= r < 3 and 0 <= c < 3):
            print("Tọa độ ngoài phạm vi. Thử lại.")
            continue
        if not valid_move(board, r, c):
            print("Ô đã có người đi. Nhập lại.")
            # nếu O cố nhập ô đã có, yêu cầu nhập lại (không đổi lượt)
            continue
        board[r, c] = current
        moves += 1
        if is_winner(board, current):
            print_board(board)
            print(f"Người {player} thắng!")
            break
        if moves == 9:
            print_board(board)
            print("Hòa.")
            break
        current = 0 if current == 1 else 1

if __name__ == "__main__":
    play()
```

```python
# BTVN 4
# tạo 1 mảng numpy có kích thước 150 x 5. Giả sử mã này chứa 150 mẫu dữ liệu của sinh viên bao gồm chiều cao, cân nặng, tuổi, lương và điểm trung bình
# tách mảng dựa trên 4 cột đầu tiên thành một biến có tên là x và cột cuối cùng là y

# tách x thành x_train, x_test trong đó x_train chứa 70% dữ liệu và tách y thành y_train, y_test trong đó y_train chứa 70% dữ liệu

# tạo 10 tập dữ liệu x_train không chồng chéo nhau
```

```python
import numpy as np

# tạo dữ liệu giả (150 x 5): [height, weight, age, salary, gpa]
rng = np.random.default_rng(42)
data = np.empty((150,5))
# ví dụ phân phối hợp lý:
data[:,0] = rng.normal(170, 10, 150)      # height (cm)
data[:,1] = rng.normal(65, 12, 150)       # weight (kg)
data[:,2] = rng.integers(18, 30, 150)     # age (yrs)
data[:,3] = rng.normal(8_000, 2_000, 150) # salary
data[:,4] = np.clip(rng.normal(3.0, 0.5, 150), 0.0, 4.0)  # gpa

# tách X và y
X = data[:, :4]
y = data[:, 4]

# trộn dữ liệu trước khi chia (giữ reproducible)
idx = rng.permutation(150)
X = X[idx]
y = y[idx]

# chia 70% train, 30% test
n_train = int(0.7 * 150)  # 105
X_train = X[:n_train]
X_test  = X[n_train:]
y_train = y[:n_train]
y_test  = y[n_train:]

# tạo 10 tập X_train không chồng chéo nhau (chia X_train thành 10 nhóm gần đều)
fold_sizes = np.full(10, n_train // 10, dtype=int)  # base size = 10
remainder = n_train % 10                              # 5
fold_sizes[:remainder] += 1                           # đầu các fold nhận 1 phần tử thêm

# tách theo fold_sizes
x_train_splits = []
start = 0
for sz in fold_sizes:
    end = start + sz
    x_train_splits.append(X_train[start:end])
    start = end

for i, split in enumerate(x_train_splits, 1):
    print(f"--- Split {i}: shape = {split.shape} ---")
    print(split[:10])
    print()
```
